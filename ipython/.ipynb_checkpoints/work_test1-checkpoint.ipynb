```python
years = range(1880,2011)
pieces = []

columns = names=['name','sex','births']

for year in years:
    path = 'names/yob%d.txt' % year
    frame = pd.read_csv(path,names = columns)
    frame['year'] = year
    pieces.append(frame)
names = pd.concat(pieces,ignore_index=True)
```
